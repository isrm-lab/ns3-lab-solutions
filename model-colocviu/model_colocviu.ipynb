```python
import copy
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
 
DATA_FILE_UDP = '/home/student/ns-3-dev/output_udp.txt.bkp'
DATA_FILE_TCP = '/home/student/ns-3-dev/output_tcp.txt.bkp'

columns = ['numberOfNodes', 'uplinkThroughput', 'downlinkThroughput']
sim_data_udp = np.genfromtxt(DATA_FILE_UDP, delimiter=' ', names=columns, dtype=None)
sim_data_tcp = np.genfromtxt(DATA_FILE_TCP, delimiter=' ', names=columns, dtype=None)


def plot_throughput_udp(sim_data):
    fig, ax = plt.subplots(figsize=(12,12))
    
    ax.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    plt.xlabel('Number of clients')
    plt.ylabel('Throughput [Mbps]')
    plt.title('UDP uplink and dowlink throughput')
 
    ax.plot(sim_data['numberOfNodes'], sim_data['uplinkThroughput'], label='Uplink')
    ax.plot(sim_data['numberOfNodes'], sim_data['downlinkThroughput'], label='Downlink')
    
    ax.legend()
    plt.savefig('whatever.png')
    plt.show()


def plot_throughput_tcp(sim_data):
    fig, ax = plt.subplots(figsize=(12,12))
    
    ax.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    plt.xlabel('Number of clients')
    plt.ylabel('Throughput [Mbps]')
    plt.title('TCP uplink and dowlink throughput')
 
    ax.plot(sim_data['numberOfNodes'], sim_data['uplinkThroughput'], label='Uplink')
    ax.plot(sim_data['numberOfNodes'], sim_data['downlinkThroughput'], label='Downlink')
    
    ax.legend()
    plt.savefig('whatever.png')
    plt.show()

    
if __name__ == '__main__':
    plot_throughput_udp(sim_data_udp)    
    plot_throughput_tcp(sim_data_tcp)
```
