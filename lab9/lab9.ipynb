```python
import copy
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

colors = [
    'blue', 'orange', 'red', 'yellow', 'black', 
    'purple', 'pink', 'brown', 'green', 'teal', 'coral', 'lightblue', 
    'lime', 'lavender', 'turquoise', 'darkgreen', 'tan', 'salmon', 'gold'
] 
matplotlib.rcParams['axes.prop_cycle'] = matplotlib.cycler(color=colors) 


DATA_FILE = '/home/student/ns-3-dev/eq_cw_fixed.txt'
DATA_FILE_STANDARD = '/home/student/ns-3-dev/eq_cw_standard.txt'
DATA_FILE_STANDARD_RTS = '/home/student/ns-3-dev/eq_cw_standard_rts.txt'

columns = ['cw', 'simulationTime', 'numberOfClients', 'jainIndex', 'epsilonIndex']
standard_columns = columns[1:]

sim_data = np.genfromtxt(DATA_FILE, delimiter=' ', names=columns)
sim_data_standard = np.genfromtxt(DATA_FILE_STANDARD, delimiter=' ', names=standard_columns)
sim_data_standard_rts = np.genfromtxt(DATA_FILE_STANDARD_RTS, delimiter=' ', names=standard_columns)


def plot_eq_metrics(sim_data):
    data = copy.deepcopy(sim_data)

    fig, ax = plt.subplots(4, 2, figsize=(15, 15))

    # Jain equity index for CW=7
    axis = ax[0][0]
    axis.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    axis.set_ylim(ymin=0.0, ymax=1.10)
    axis.set_xlabel('Number of clients')
    axis.set_ylabel('Equity Index')
    axis.set_title('Jain Equity Index for CW=%s' % data[19 * 0]['cw'])

    axis.plot(data['numberOfClients'][19 * 0:19 * 1], data['jainIndex'][19 * 0:19 * 1], label=int(data['simulationTime'][19 * 0]))
    axis.plot(data['numberOfClients'][19 * 1:19 * 2], data['jainIndex'][19 * 1:19 * 2], label=int(data['simulationTime'][19 * 1]))
    axis.plot(data['numberOfClients'][19 * 2:19 * 3], data['jainIndex'][19 * 2:19 * 3], label=int(data['simulationTime'][19 * 2]))
    axis.legend()

    # Epsilon equity index for CW=7
    axis = ax[0][1]
    axis.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    axis.set_ylim(ymin=0.0, ymax=1.10)
    axis.set_xlabel('Number of clients')
    axis.set_ylabel('Equity Index')
    axis.set_title('Epsilon Equity Index for CW=%s' % data[19 * 0]['cw'])

    axis.plot(data['numberOfClients'][19 * 0:19 * 1], data['epsilonIndex'][19 * 0:19 * 1], label=int(data['simulationTime'][19 * 0]))
    axis.plot(data['numberOfClients'][19 * 1:19 * 2], data['epsilonIndex'][19 * 1:19 * 2], label=int(data['simulationTime'][19 * 1]))
    axis.plot(data['numberOfClients'][19 * 2:19 * 3], data['epsilonIndex'][19 * 2:19 * 3], label=int(data['simulationTime'][19 * 2]))
    axis.legend()   
  
    # Jain equity index for CW=31
    axis = ax[1][0]
    axis.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    axis.set_ylim(ymin=0.0, ymax=1.10)
    axis.set_xlabel('Number of clients')
    axis.set_ylabel('Equity Index')
    axis.set_title('Jain Equity Index for CW=%s' % data[19 * 3]['cw'])

    axis.plot(data['numberOfClients'][19 * 3:19 * 4], data['jainIndex'][19 * 3:19 * 4], label=int(data['simulationTime'][19 * 3]))
    axis.plot(data['numberOfClients'][19 * 4:19 * 5], data['jainIndex'][19 * 4:19 * 5], label=int(data['simulationTime'][19 * 4]))
    axis.plot(data['numberOfClients'][19 * 5:19 * 6], data['jainIndex'][19 * 5:19 * 6], label=int(data['simulationTime'][19 * 5]))
    axis.legend()

    # Epsilon equity index for CW=31
    axis = ax[1][1]
    axis.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    axis.set_ylim(ymin=0.0, ymax=1.10)
    axis.set_xlabel('Number of clients')
    axis.set_ylabel('Equity Index')
    axis.set_title('Epsilon Equity Index for CW=%s' % sim_data[19 * 3]['cw'])

    axis.plot(data['numberOfClients'][19 * 3:19 * 4], data['epsilonIndex'][19 * 3:19 * 4], label=int(data['simulationTime'][19 * 3]))
    axis.plot(data['numberOfClients'][19 * 4:19 * 5], data['epsilonIndex'][19 * 4:19 * 5], label=int(data['simulationTime'][19 * 4]))
    axis.plot(data['numberOfClients'][19 * 5:19 * 6], data['epsilonIndex'][19 * 5:19 * 6], label=int(data['simulationTime'][19 * 5]))
    axis.legend()  

    # Jain equity index for CW=511
    axis = ax[2][0]
    axis.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    axis.set_ylim(ymin=0.0, ymax=1.10)
    axis.set_xlabel('Number of clients')
    axis.set_ylabel('Equity Index')
    axis.set_title('Jain Equity Index for CW=%s' % data[19 * 6]['cw'])

    axis.plot(data['numberOfClients'][19 * 6:19 * 7], data['jainIndex'][19 * 6:19 * 7], label=int(data['simulationTime'][19 * 6]))
    axis.plot(data['numberOfClients'][19 * 7:19 * 8], data['jainIndex'][19 * 7:19 * 8], label=int(data['simulationTime'][19 * 7]))
    axis.plot(data['numberOfClients'][19 * 8:19 * 9], data['jainIndex'][19 * 8:19 * 9], label=int(data['simulationTime'][19 * 8]))
    axis.legend()

    # Epsilon equity index for CW=511
    axis = ax[2][1]
    axis.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    axis.set_ylim(ymin=0.0, ymax=1.10)
    axis.set_xlabel('Number of clients')
    axis.set_ylabel('Equity Index')
    axis.set_title('Epsilon Equity Index for CW=%s' % sim_data[19 * 6]['cw'])

    axis.plot(data['numberOfClients'][19 * 6:19 * 7], data['epsilonIndex'][19 * 6:19 * 7], label=int(data['simulationTime'][19 * 6]))
    axis.plot(data['numberOfClients'][19 * 7:19 * 8], data['epsilonIndex'][19 * 7:19 * 8], label=int(data['simulationTime'][19 * 7]))
    axis.plot(data['numberOfClients'][19 * 8:19 * 9], data['epsilonIndex'][19 * 8:19 * 9], label=int(data['simulationTime'][19 * 8]))
    axis.legend()

    # Jain equity index for CW=4095
    axis = ax[3][0]
    axis.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    axis.set_ylim(ymin=0.0, ymax=1.10)
    axis.set_xlabel('Number of clients')
    axis.set_ylabel('Equity Index')
    axis.set_title('Jain Equity Index for CW=%s' % data[19 * 9]['cw'])

    axis.plot(data['numberOfClients'][19 * 9:19 * 10], data['jainIndex'][19 * 9:19 * 10], label=int(data['simulationTime'][19 * 9]))
    axis.plot(data['numberOfClients'][19 * 10:19 * 11], data['jainIndex'][19 * 10:19 * 11], label=int(data['simulationTime'][19 * 10]))
    axis.plot(data['numberOfClients'][19 * 11:19 * 12], data['jainIndex'][19 * 11:19 * 12], label=int(data['simulationTime'][19 * 11]))
    axis.legend()

    # Epsilon equity index for CW=4095
    axis = ax[3][1]
    axis.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    axis.set_ylim(ymin=0.0, ymax=1.10)
    axis.set_xlabel('Number of clients')
    axis.set_ylabel('Equity Index')
    axis.set_title('Epsilon Equity Index for CW=%s' % sim_data[19 * 9]['cw'])

    axis.plot(data['numberOfClients'][19 * 9:19 * 10], data['epsilonIndex'][19 * 9:19 * 10], label=int(data['simulationTime'][19 * 9]))
    axis.plot(data['numberOfClients'][19 * 10:19 * 11], data['epsilonIndex'][19 * 10:19 * 11], label=int(data['simulationTime'][19 * 10]))
    axis.plot(data['numberOfClients'][19 * 11:19 * 12], data['epsilonIndex'][19 * 11:19 * 12], label=int(data['simulationTime'][19 * 11]))
    axis.legend()

    fig.subplots_adjust(hspace=0.5)
    figure_name = 'eq_metrics.png'
    plt.savefig(figure_name)
    plt.show()


def plot_eq_metrics_standard_and_rts(sim_data_standard, sim_data_standard_rts):
    data = copy.deepcopy(sim_data_standard)

    fig, ax = plt.subplots(2, 2, figsize=(15, 15))

    # Jain equity index for standard
    axis = ax[0][0]
    axis.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    axis.set_ylim(ymin=0.0, ymax=1.10)
    axis.set_xlabel('Number of clients')
    axis.set_ylabel('Equity Index')
    axis.set_title('Jain Equity Index for standard')

    axis.plot(data['numberOfClients'][19 * 0:19 * 1], data['jainIndex'][19 * 0:19 * 1], label=int(data['simulationTime'][19 * 0]))
    axis.plot(data['numberOfClients'][19 * 1:19 * 2], data['jainIndex'][19 * 1:19 * 2], label=int(data['simulationTime'][19 * 1]))
    axis.plot(data['numberOfClients'][19 * 2:19 * 3], data['jainIndex'][19 * 2:19 * 3], label=int(data['simulationTime'][19 * 2]))
    axis.legend()

    # Epsilon equity index for standard
    axis = ax[0][1]
    axis.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    axis.set_ylim(ymin=0.0, ymax=1.10)
    axis.set_xlabel('Number of clients')
    axis.set_ylabel('Equity Index')
    axis.set_title('Epsilon Equity Index for standard')

    axis.plot(data['numberOfClients'][19 * 0:19 * 1], data['epsilonIndex'][19 * 0:19 * 1], label=int(data['simulationTime'][19 * 0]))
    axis.plot(data['numberOfClients'][19 * 1:19 * 2], data['epsilonIndex'][19 * 1:19 * 2], label=int(data['simulationTime'][19 * 1]))
    axis.plot(data['numberOfClients'][19 * 2:19 * 3], data['epsilonIndex'][19 * 2:19 * 3], label=int(data['simulationTime'][19 * 2]))
    axis.legend()

    data = copy.deepcopy(sim_data_standard_rts)
    # Jain equity index for standard with RTS/CTS
    axis = ax[1][0]
    axis.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    axis.set_ylim(ymin=0.0, ymax=1.10)
    axis.set_xlabel('Number of clients')
    axis.set_ylabel('Equity Index')
    axis.set_title('Jain Equity Index for standard with RTS/CTS')

    axis.plot(data['numberOfClients'][19 * 0:19 * 1], data['jainIndex'][19 * 0:19 * 1], label=int(data['simulationTime'][19 * 0]))
    axis.plot(data['numberOfClients'][19 * 1:19 * 2], data['jainIndex'][19 * 1:19 * 2], label=int(data['simulationTime'][19 * 1]))
    axis.plot(data['numberOfClients'][19 * 2:19 * 3], data['jainIndex'][19 * 2:19 * 3], label=int(data['simulationTime'][19 * 2]))
    axis.legend()

    # Epsilon equity index for standard with RTS/CTS
    axis = ax[1][1]
    axis.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    axis.set_ylim(ymin=0.0, ymax=1.10)
    axis.set_xlabel('Number of clients')
    axis.set_ylabel('Equity Index')
    axis.set_title('Epsilon Equity Index for standard with RTS/CTS')

    axis.plot(data['numberOfClients'][19 * 0:19 * 1], data['epsilonIndex'][19 * 0:19 * 1], label=int(data['simulationTime'][19 * 0]))
    axis.plot(data['numberOfClients'][19 * 1:19 * 2], data['epsilonIndex'][19 * 1:19 * 2], label=int(data['simulationTime'][19 * 1]))
    axis.plot(data['numberOfClients'][19 * 2:19 * 3], data['epsilonIndex'][19 * 2:19 * 3], label=int(data['simulationTime'][19 * 2]))
    axis.legend()

    fig.subplots_adjust(hspace=0.3)
    figure_name = 'eq_metrics_standard_and_rts.png'
    plt.savefig(figure_name)
    plt.show()

if __name__ == '__main__':
    plot_eq_metrics(sim_data)
    plot_eq_metrics_standard_and_rts(sim_data_standard, sim_data_standard_rts) 
```
