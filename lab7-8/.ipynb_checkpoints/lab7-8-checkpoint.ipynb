```python
import copy
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

DATA_FILE = '/home/student/ns-3-dev/cw_stats.txt'
columns = ['sz', 'cw', 'sent_agt', 'recv_agt', 'col_cbr', 'sent_mac', 'recv_mac', 'avgTput']
sim_data = np.genfromtxt(DATA_FILE, delimiter=' ', skip_header=1, names=columns)


def plot_metrics(sim_data):
    data = copy.deepcopy(sim_data)

    fig, ax = plt.subplots(4, 2, figsize=(20,20))

    # Total pachete CBR livrate
    ax[0][0].set_xscale('log')
    ax[0][0].grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    ax[0][0].set_xlabel('Contention Window slots')
    ax[0][0].set_ylabel('Total pachete CBR livrate')

    ax[0][0].plot(data['cw'][:11], data['recv_agt'][:11], label=int(data['sz'][0]))
    ax[0][0].plot(data['cw'][11:22], data['recv_agt'][11:22], label=int(data['sz'][11]))
    ax[0][0].plot(data['cw'][22:33], data['recv_agt'][22:33], label=int(data['sz'][22]))
    ax[0][0].plot(data['cw'][33:44], data['recv_agt'][33:44], label=int(data['sz'][33]))
    ax[0][0].plot(data['cw'][44:55], data['recv_agt'][44:55], label=int(data['sz'][44]))
    ax[0][0].legend()

    # Debit la livrare
    ax[0][1].set_xscale('log')
    ax[0][1].grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    ax[0][1].set_xlabel('Contention Window slots')
    ax[0][1].set_ylabel('Debit la livrare [Mbps]')

    data['recv_agt'] = [data_point * 1512 * 8 / (25 * 10**6) for data_point in data['recv_agt']]
    ax[0][1].plot(data['cw'][:11], data['recv_agt'][:11], label=int(data['sz'][0]))
    ax[0][1].plot(data['cw'][11:22], data['recv_agt'][11:22], label=int(data['sz'][11]))
    ax[0][1].plot(data['cw'][22:33], data['recv_agt'][22:33], label=int(data['sz'][22]))
    ax[0][1].plot(data['cw'][33:44], data['recv_agt'][33:44], label=int(data['sz'][33]))
    ax[0][1].plot(data['cw'][44:55], data['recv_agt'][44:55], label=int(data['sz'][44]))
    ax[0][1].legend()

    # PDR UDP
    ax[1][0].set_xscale('log')
    ax[1][0].grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    ax[1][0].set_xlabel('Contention Window slots')
    ax[1][0].set_ylabel('PDR UDP')

    data_points = [recv_agt / sent_agt for recv_agt, sent_agt in zip(data['recv_agt'], data['sent_agt'])]
    ax[1][0].plot(data['cw'][:11], data_points[:11], label=int(data['sz'][0]))
    ax[1][0].plot(data['cw'][11:22], data_points[11:22], label=int(data['sz'][11]))
    ax[1][0].plot(data['cw'][22:33], data_points[22:33], label=int(data['sz'][22]))
    ax[1][0].plot(data['cw'][33:44], data_points[33:44], label=int(data['sz'][33]))
    ax[1][0].plot(data['cw'][44:55], data_points[44:55], label=int(data['sz'][44]))
    ax[1][0].legend()

    # PDR MAC
    ax[1][1].set_xscale('log')
    ax[1][1].grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    ax[1][1].set_xlabel('Contention Window slots')
    ax[1][1].set_ylabel('PDR MAC')

    data_points = [recv_agt / sent_mac for recv_agt, sent_mac in zip(data['recv_agt'], data['sent_mac'])]
    ax[1][1].plot(data['cw'][:11], data_points[:11], label=int(data['sz'][0]))
    ax[1][1].plot(data['cw'][11:22], data_points[11:22], label=int(data['sz'][11]))
    ax[1][1].plot(data['cw'][22:33], data_points[22:33], label=int(data['sz'][22]))
    ax[1][1].plot(data['cw'][33:44], data_points[33:44], label=int(data['sz'][33]))
    ax[1][1].plot(data['cw'][44:55], data_points[44:55], label=int(data['sz'][44]))
    ax[1][1].legend()

    # Pachete UDP emise de agent
    ax[2][0].set_xscale('log')
    ax[2][0].grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    ax[2][0].set_xlabel('Contention Window slots')
    ax[2][0].set_ylabel('Pachete UDP emise de agent')

    ax[2][0].plot(data['cw'][:11], data['sent_agt'][:11], label=int(data['sz'][0]))
    ax[2][0].plot(data['cw'][11:22], data['sent_agt'][11:22], label=int(data['sz'][11]))
    ax[2][0].plot(data['cw'][22:33], data['sent_agt'][22:33], label=int(data['sz'][22]))
    ax[2][0].plot(data['cw'][33:44], data['sent_agt'][33:44], label=int(data['sz'][33]))
    ax[2][0].plot(data['cw'][44:55], data['sent_agt'][44:55], label=int(data['sz'][44]))
    ax[2][0].legend()

    # Pachete UDP emise de MAC
    ax[2][1].set_xscale('log')
    ax[2][1].grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    ax[2][1].set_xlabel('Contention Window slots')
    ax[2][1].set_ylabel('Pachete UDP emise de MAC')

    ax[2][1].plot(data['cw'][:11], data['sent_mac'][:11], label=int(data['sz'][0]))
    ax[2][1].plot(data['cw'][11:22], data['sent_mac'][11:22], label=int(data['sz'][11]))
    ax[2][1].plot(data['cw'][22:33], data['sent_mac'][22:33], label=int(data['sz'][22]))
    ax[2][1].plot(data['cw'][33:44], data['sent_mac'][33:44], label=int(data['sz'][33]))
    ax[2][1].plot(data['cw'][44:55], data['sent_mac'][44:55], label=int(data['sz'][44]))
    ax[2][1].legend()

    # Numar coliziuni la nivel de date
    ax[3][0].set_xscale('log')
    ax[3][0].grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    ax[3][0].set_xlabel('Contention Window slots')
    ax[3][0].set_ylabel('Numar coliziuni date')

    ax[3][0].plot(data['cw'][:11], data['col_cbr'][:11], label=int(data['sz'][0]))
    ax[3][0].plot(data['cw'][11:22], data['col_cbr'][11:22], label=int(data['sz'][11]))
    ax[3][0].plot(data['cw'][22:33], data['col_cbr'][22:33], label=int(data['sz'][22]))
    ax[3][0].plot(data['cw'][33:44], data['col_cbr'][33:44], label=int(data['sz'][33]))
    ax[3][0].plot(data['cw'][44:55], data['col_cbr'][44:55], label=int(data['sz'][44]))
    ax[3][0].legend()

    # Transmisiuni per pachet
    ax[3][1].set_xscale('log')
    ax[3][1].grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    ax[3][1].set_xlabel('Contention Window slots')
    ax[3][1].set_ylabel('Transmisiuni per pachet')

    data_points = [sent_mac / recv_agt for sent_mac, recv_agt in zip(data['sent_mac'], data['recv_agt'])]
    ax[3][1].plot(data['cw'][:11], data_points[:11], label=int(data['sz'][0]))
    ax[3][1].plot(data['cw'][11:22], data_points[11:22], label=int(data['sz'][11]))
    ax[3][1].plot(data['cw'][22:33], data_points[22:33], label=int(data['sz'][22]))
    ax[3][1].plot(data['cw'][33:44], data_points[33:44], label=int(data['sz'][33]))
    ax[3][1].plot(data['cw'][44:55], data_points[44:55], label=int(data['sz'][44]))
    ax[3][1].legend()
    
    fig.subplots_adjust(hspace=0.5)
    fig.subplots_adjust(wspace=0.5)
    plt.show()

def plot_total_cbr_packets_delivered(sim_data):
    data = copy.deepcopy(sim_data)

    fig, ax = plt.subplots(figsize=(12,12))
    ax.set_xscale('log')
    ax.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    plt.xlabel('Contention Window slots')
    plt.ylabel('Total pachete CBR livrate')

    ax.plot(data['cw'][:11], data['recv_agt'][:11], label=int(data['sz'][0]))
    ax.plot(data['cw'][11:22], data['recv_agt'][11:22], label=int(data['sz'][11]))
    ax.plot(data['cw'][22:33], data['recv_agt'][22:33], label=int(data['sz'][22]))
    ax.plot(data['cw'][33:44], data['recv_agt'][33:44], label=int(data['sz'][33]))
    ax.plot(data['cw'][44:55], data['recv_agt'][44:55], label=int(data['sz'][44]))
    ax.legend()

    plt.show()

# def plot_rate_delivered(sim_data):
#     data = copy.deepcopy(sim_data)
#     fig, ax = plt.subplots(figsize=(12,12))
#     ax.set_xscale('log')
#     ax.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
#     plt.xlabel('Contention Window slots')
#     plt.ylabel('Debit la livrare [Mbps]')

#     data['recv_agt'] = [data_point * 1512 * 8 / (25 * 10**6) for data_point in data['recv_agt']]
#     ax.plot(data['cw'][:11], data['recv_agt'][:11], label=int(data['sz'][0]))
#     ax.plot(data['cw'][11:22], data['recv_agt'][11:22], label=int(data['sz'][11]))
#     ax.plot(data['cw'][22:33], data['recv_agt'][22:33], label=int(data['sz'][22]))
#     ax.plot(data['cw'][33:44], data['recv_agt'][33:44], label=int(data['sz'][33]))
#     ax.plot(data['cw'][44:55], data['recv_agt'][44:55], label=int(data['sz'][44]))
#     ax.legend()

#     plt.show()

# def plot_cbr_packet_delivery_ratio(sim_data):
#     data = copy.deepcopy(sim_data)
#     fig, ax = plt.subplots(figsize=(12,12))
#     ax.set_xscale('log')
#     ax.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
#     plt.xlabel('Contention Window slots')
#     plt.ylabel('Debit la livrare [Mbps]')

#     data_points = [recv_agt / sent_agt for recv_agt, sent_agt in zip(data['recv_agt'], data['sent_agt'])]
#     ax.plot(data['cw'][:11], data_points[:11], label=int(data['sz'][0]))
#     ax.plot(data['cw'][11:22], data_points[11:22], label=int(data['sz'][11]))
#     ax.plot(data['cw'][22:33], data_points[22:33], label=int(data['sz'][22]))
#     ax.plot(data['cw'][33:44], data_points[33:44], label=int(data['sz'][33]))
#     ax.plot(data['cw'][44:55], data_points[44:55], label=int(data['sz'][44]))
#     ax.legend()

#     plt.show()

    
# def plot_mac_packet_delivery_ratio(sim_data):
#     data = copy.deepcopy(sim_data)
#     fig, ax = plt.subplots(figsize=(12,12))
#     ax.set_xscale('log')
#     ax.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
#     plt.xlabel('Contention Window slots')
#     plt.ylabel('Debit la livrare [Mbps]')

#     data_points = [recv_agt / sent_mac for recv_agt, sent_mac in zip(data['recv_agt'], data['sent_mac'])]
#     ax.plot(data['cw'][:11], data_points[:11], label=int(data['sz'][0]))
#     ax.plot(data['cw'][11:22], data_points[11:22], label=int(data['sz'][11]))
#     ax.plot(data['cw'][22:33], data_points[22:33], label=int(data['sz'][22]))
#     ax.plot(data['cw'][33:44], data_points[33:44], label=int(data['sz'][33]))
#     ax.plot(data['cw'][44:55], data_points[44:55], label=int(data['sz'][44]))
#     ax.legend()

#     plt.show()


# def plot_number_of_cbr_packets_sent_by_agent(sim_data):
#     data = copy.deepcopy(sim_data)

#     fig, ax = plt.subplots(figsize=(12,12))
#     ax.set_xscale('log')
#     ax.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
#     plt.xlabel('Contention Window slots')
#     plt.ylabel('Pachete CBR emise de agent')

#     ax.plot(data['cw'][:11], data['sent_agt'][:11], label=int(data['sz'][0]))
#     ax.plot(data['cw'][11:22], data['sent_agt'][11:22], label=int(data['sz'][11]))
#     ax.plot(data['cw'][22:33], data['sent_agt'][22:33], label=int(data['sz'][22]))
#     ax.plot(data['cw'][33:44], data['sent_agt'][33:44], label=int(data['sz'][33]))
#     ax.plot(data['cw'][44:55], data['sent_agt'][44:55], label=int(data['sz'][44]))
#     ax.legend()

#     plt.show()

# def plot_number_of_cbr_packets_sent_by_mac(sim_data):
#     data = copy.deepcopy(sim_data)

#     fig, ax = plt.subplots(figsize=(12,12))
#     ax.set_xscale('log')
#     ax.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
#     plt.xlabel('Contention Window slots')
#     plt.ylabel('Pachete CBR emise de MAC')

#     ax.plot(data['cw'][:11], data['sent_mac'][:11], label=int(data['sz'][0]))
#     ax.plot(data['cw'][11:22], data['sent_mac'][11:22], label=int(data['sz'][11]))
#     ax.plot(data['cw'][22:33], data['sent_mac'][22:33], label=int(data['sz'][22]))
#     ax.plot(data['cw'][33:44], data['sent_mac'][33:44], label=int(data['sz'][33]))
#     ax.plot(data['cw'][44:55], data['sent_mac'][44:55], label=int(data['sz'][44]))
#     ax.legend()

#     plt.show()

# def plot_number_of_collisions_per_second(sim_data):
#     data = copy.deepcopy(sim_data)

#     fig, ax = plt.subplots(figsize=(12,12))
#     ax.set_xscale('log')
#     ax.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
#     plt.xlabel('Contention Window slots')
#     plt.ylabel('Numar coliziuni date')

#     ax.plot(data['cw'][:11], data['col_cbr'][:11], label=int(data['sz'][0]))
#     ax.plot(data['cw'][11:22], data['col_cbr'][11:22], label=int(data['sz'][11]))
#     ax.plot(data['cw'][22:33], data['col_cbr'][22:33], label=int(data['sz'][22]))
#     ax.plot(data['cw'][33:44], data['col_cbr'][33:44], label=int(data['sz'][33]))
#     ax.plot(data['cw'][44:55], data['col_cbr'][44:55], label=int(data['sz'][44]))
#     ax.legend()

#     plt.show()

    
# def plot_transmissions_per_packet(sim_data):
#     data = copy.deepcopy(sim_data)
#     fig, ax = plt.subplots(figsize=(12,12))
#     ax.set_xscale('log')
#     ax.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
#     plt.xlabel('Contention Window slots')
#     plt.ylabel('Transmisiuni per pachet')

#     data_points = [sent_mac / recv_agt for sent_mac, recv_agt in zip(data['sent_mac'], data['recv_agt'])]
#     ax.plot(data['cw'][:11], data_points[:11], label=int(data['sz'][0]))
#     ax.plot(data['cw'][11:22], data_points[11:22], label=int(data['sz'][11]))
#     ax.plot(data['cw'][22:33], data_points[22:33], label=int(data['sz'][22]))
#     ax.plot(data['cw'][33:44], data_points[33:44], label=int(data['sz'][33]))
#     ax.plot(data['cw'][44:55], data_points[44:55], label=int(data['sz'][44]))
#     ax.legend()

#     plt.show()

    
if __name__ == '__main__':
#     plot_metrics(sim_data)
#     plot_total_cbr_packets_delivered(sim_data)
#     plot_rate_delivered(sim_data)
#     plot_cbr_packet_delivery_ratio(sim_data)
#     plot_mac_packet_delivery_ratio(sim_data)
#     plot_number_of_cbr_packets_sent_by_agent(sim_data)
#     plot_number_of_cbr_packets_sent_by_mac(sim_data)
#     plot_number_of_collisions_per_second(sim_data)
#     plot_transmissions_per_packet(sim_data)
```
