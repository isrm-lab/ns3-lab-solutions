```python
import copy
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
 
DATA_FILE_802_11_B_UDP_RTS_CTS = '/home/student/ns-3-dev/lab4_1_80211_b.txt'
DATA_FILE_802_11_B_TCP = '/home/student/ns-3-dev/lab4_2_80211_b.txt'
DATA_FILE_802_11_G_TCP = '/home/student/ns-3-dev/lab4_2_80211_g.txt'
DATA_FILE_802_11_B_TCP_RTS = '/home/student/ns-3-dev/lab4_3_80211_b.txt'
DATA_FILE_802_11_G_TCP_RTS = '/home/student/ns-3-dev/lab4_3_80211_g.txt'
DATA_FILE_UPLINK_UDP = '/home/student/ns-3-dev/lab4_4_udp.txt'
DATA_FILE_UPLINK_TCP = '/home/student/ns-3-dev/lab4_4_tcp.txt'

columns = ['mcs', 'payload_size', 'throughput']
uplink_columns = ['number_of_nodes', 'throughput']

sim_data_802_11_b_udp_rts_cts = np.genfromtxt(DATA_FILE_802_11_B_UDP_RTS_CTS, delimiter=' ', names=columns, dtype=None)
sim_data_802_11_b_tcp = np.genfromtxt(DATA_FILE_802_11_B_TCP, delimiter=' ', names=columns, dtype=None)
sim_data_802_11_g_tcp = np.genfromtxt(DATA_FILE_802_11_G_TCP, delimiter=' ', names=columns, dtype=None)
sim_data_802_11_b_tcp_rts = np.genfromtxt(DATA_FILE_802_11_B_TCP_RTS, delimiter=' ', names=columns, dtype=None)
sim_data_802_11_g_tcp_rts = np.genfromtxt(DATA_FILE_802_11_G_TCP_RTS, delimiter=' ', names=columns, dtype=None)

sim_data_udp = np.genfromtxt(DATA_FILE_UPLINK_UDP, delimiter=' ', names=uplink_columns, dtype=None)
sim_data_tcp = np.genfromtxt(DATA_FILE_UPLINK_TCP, delimiter=' ', names=uplink_columns, dtype=None)

def plot_simulation_802_11_b_udp_rts_cts_throughput(sim_data):
    fig, ax = plt.subplots(figsize=(12,12))
    ax.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
 
    plt.xlabel('Payload size [bytes]')
    plt.ylabel('Throughput [Mbps]')
    plt.title('Simulation 802.11b UDP RTS/CTS throughput')
 
    for i in range(0,len(sim_data), 6):
        ax.plot(sim_data['payload_size'][i:i + 6], sim_data['throughput'][i:i + 6], label=sim_data['mcs'][i])
    ax.legend()
    plt.savefig("simulation_80211_b_udp_rts_cts.png")
    plt.show()

def plot_simulation_802_11_b_tcp_throughput(sim_data):
    fig, ax = plt.subplots(figsize=(12,12))
    ax.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
 
    plt.xlabel('Payload size [bytes]')
    plt.ylabel('Throughput [Mbps]')
    plt.title('Simulation 802.11b TCP throughput')
 
    for i in range(0,len(sim_data), 6):
        ax.plot(sim_data['payload_size'][i:i + 6], sim_data['throughput'][i:i + 6], label=sim_data['mcs'][i])
    ax.legend()
    plt.savefig("simulation_80211_b_tcp.png")
    plt.show()

def plot_simulation_802_11_g_tcp_throughput(sim_data):
    fig, ax = plt.subplots(figsize=(12,12))
    ax.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
 
    plt.xlabel('Payload size [bytes]')
    plt.ylabel('Throughput [Mbps]')
    plt.title('Simulation 802.11g TCP throughput')
 
    for i in range(0,len(sim_data), 6):
        ax.plot(sim_data['payload_size'][i:i + 6], sim_data['throughput'][i:i + 6], label=sim_data['mcs'][i])
    ax.legend()
    plt.savefig("simulation_80211_g_tcp.png")
    plt.show()

def plot_simulation_802_11_b_tcp_rts_throughput(sim_data):
    fig, ax = plt.subplots(figsize=(12,12))
    ax.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
 
    plt.xlabel('Payload size [bytes]')
    plt.ylabel('Throughput [Mbps]')
    plt.title('Simulation 802.11b TCP RTS throughput')
 
    for i in range(0,len(sim_data), 6):
        ax.plot(sim_data['payload_size'][i:i + 6], sim_data['throughput'][i:i + 6], label=sim_data['mcs'][i])
    ax.legend()
    plt.savefig("simulation_80211_b_tcp_rts.png")
    plt.show()

def plot_simulation_802_11_g_tcp_rts_throughput(sim_data):
    fig, ax = plt.subplots(figsize=(12,12))
    ax.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
 
    plt.xlabel('Payload size [bytes]')
    plt.ylabel('Throughput [Mbps]')
    plt.title('Simulation 802.11g TCP RTS throughput')
 
    for i in range(0,len(sim_data), 6):
        ax.plot(sim_data['payload_size'][i:i + 6], sim_data['throughput'][i:i + 6], label=sim_data['mcs'][i])
    ax.legend()
    plt.savefig("simulation_80211_g_tcp_rts.png")
    plt.show()

def plot_simulation_uplink_udp_throughput(sim_data):
    fig, ax = plt.subplots(figsize=(12,12))
    ax.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
 
    plt.xlabel('Number of clients')
    plt.ylabel('Throughput [Mbps]')
    plt.title('Simulation UDP uplink throughput')
 
    ax.set_ylim(bottom=0, top=11)
    ax.plot(sim_data['number_of_nodes'], sim_data['throughput'], label='UDP uplink')
    ax.legend()
    plt.savefig("simulation_uplink_udp.png")
    plt.show()

def plot_simulation_uplink_tcp_throughput(sim_data):
    fig, ax = plt.subplots(figsize=(12,12))
    ax.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
 
    plt.xlabel('Number of clients')
    plt.ylabel('Throughput [Mbps]')
    plt.title('Simulation TCP uplink throughput')
 
    ax.plot(sim_data['number_of_nodes'], sim_data['throughput'], label='TCP uplink')
    ax.legend()
    plt.savefig("simulation_uplink_tcp.png")
    plt.show()

if __name__ == '__main__':
    plot_simulation_802_11_b_udp_rts_cts_throughput(sim_data_802_11_b_udp_rts_cts)    
    plot_simulation_802_11_b_tcp_throughput(sim_data_802_11_b_tcp)
    plot_simulation_802_11_g_tcp_throughput(sim_data_802_11_g_tcp)
    plot_simulation_802_11_b_tcp_rts_throughput(sim_data_802_11_b_tcp_rts)
    plot_simulation_802_11_g_tcp_rts_throughput(sim_data_802_11_g_tcp_rts)
    plot_simulation_uplink_udp_throughput(sim_data_udp)
    plot_simulation_uplink_tcp_throughput(sim_data_tcp)
```
